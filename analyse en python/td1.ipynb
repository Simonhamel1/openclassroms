```python
from math import *

def methode_fixe(fct, epsilon, Max):
    n = 0
    xn = 4 # a changer selon la fonction
    xn1 = fct(xn)
    while abs(xn1 - xn)/abs(xn1) > epsilon and n < Max:
        xn, xn1 = xn1, fct(xn1)
        n += 1
    return xn1, n

```

```python
fct1 = lambda x: sqrt(2 * x + 3)

solution, iterations = methode_fixe(fct1, 0.00000001, 100)
print("Solution approchée :", solution)
print("Nombre d'itérations :", iterations)
```

```python
fct1 = lambda x: 3 / (x - 2)
solution, iterations = methode_fixe(fct1, 0.00000001, 100)
print("Solution approchée :", solution)
print("Nombre d'itérations :", iterations)
```

```python
fct1 = lambda x: (x**2 - 3) / 2

solution, iterations = methode_fixe(fct1, 0.00000001, 1000)
print("Solution approchée :", solution)
print("Nombre d'itérations :", iterations)
```

# Exo 2

```python
def methode_newton(fct, dfct, epsilon, Max):
    n = 0
    xn = 1 # a changer selon la fonction
    xn1 = xn - fct(xn) / dfct(xn)
    while abs(xn1 - xn)/abs(xn1) > epsilon and n < Max:
        xn, xn1 = xn1, xn1 - fct(xn1) / dfct(xn1)
        n += 1
    return xn1, n
```

```python
fct = lambda x: x**2 - 2
dfct = lambda x: 2 * x

print("\nMéthode de Newton-Raphson pour f(x) = x^2 - 2")
solution, iterations = methode_newton(fct, dfct, 0.00000001, 100)
print("Solution approchée :", solution)
print("Nombre d'itérations :", iterations)
```
