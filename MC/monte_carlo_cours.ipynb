```python
import numpy as np
import matplotlib.pyplot as plt

def bernou(p):
    u = np.random.rand()
    if u < p:
        return 1
    else:
        return 0


def EchBern(N, p):
    result = []
    for i in range(N):
        result.append(bernou(p))
    return result


# paramètres
N = 100
p = 0.3

# échantillon
test = EchBern(N, p)
moy = np.mean(test)

# comptage des 0 et des 1
valeurs = [0, 1]
effectifs = [test.count(0), test.count(1)]
bartheo =[1-p,p]
baremp=[1-moy, moy]
# faut les valeurs empiriques et les moyennes théoriques


# graphe en bâtons
plt.bar([0,1], bartheo, align='edge', width=0.2, alpha=0.5, color='r')
plt.bar([0,1], baremp, align='edge', width=-0.2, alpha=0.5, color='b')
# titres et labels
plt.title("Loi de Bernoulli (p = 0.3)")
plt.xlabel("Valeurs")
plt.ylabel("Effectifs")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# simulation d'une v.a. discrète prenant les valeurs ai avec des proba pi
# val = valeurs prises
# proba = probabilités

def discret(val, proba):
    u = np.random.random()
    F = proba[0]
    x = val[0]
    k = 0

    while u > F:
        k += 1
        F += proba[k]
        x = val[k]

    return x


# génération échantillon discret
def EchDiscr(N, val, proba):
    Ech = []
    for i in range(N):
        Ech.append(discret(val, proba))
    return Ech


N = 1000
val = np.array([1, 2, 3, 4, 5, 6])
proba = np.array([0.2, 0.15, 0.22, 0.09, 0.18, 0.16])

echN = EchDiscr(N, val, proba)

# calcul espérance et variance théoriques
esptheo = sum(val * proba)
vartheo = sum(val**2 * proba) - esptheo**2

# calcul espérance et variance empiriques
espempi = np.mean(echN)
varempi = np.var(echN)

# probabilités empiriques
probaEmp = []
for i in range(len(val)):
    probaEmp.append(echN.count(val[i]) / N)


# représentation graphique
plt.bar(val, proba, align='edge', width=0.2, alpha=0.5, color='r')
plt.bar(val, probaEmp, align='edge', width=-0.2, alpha=0.5, color='b')

plt.show()
```
