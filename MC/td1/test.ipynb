## Loi de Poisson :

La loi de Poisson est une loi de probabilité discrète qui modélise 
le nombre d'événements se produisant dans un intervalle de temps ou d'espace donné,
lorsque ces événements se produisent de manière indépendante et à un taux constant.
Elle est souvent utilisée pour modéliser des phénomènes tels que
le nombre d'appels reçus par un centre d'appels en une heure,
le nombre de voitures passant par un péage en une journée,
ou le nombre de défauts dans une pièce manufacturée.

```python
from math import exp

def proba_poisson_inferieur(lam, k):
    #p(X = 0)
    current_term = exp(-lam)
    cumulative_proba = current_term

    for i in range(1, k + 1):
        current_term *= lam / i
        cumulative_proba += current_term
        
    return cumulative_proba
```

```python
def proba_poisson_egal(lam, k):
    current_term = exp(-lam)
    cumulative_proba = current_term
    
    for i in range(1, k + 1):
        current_term *= lam / i
        if i == k:
            return current_term
    
    return cumulative_proba
```

```python
print("la proba d'avoir ", proba_poisson_inferieur(100, 100))
print("la proba d'avoir ", proba_poisson_egal(100, 21))
```

```python
for i in [5, 10, 15, 20]:
    tab1 = [proba_poisson_egal(i, k) for k in range(0, 30)]
    tab2 = [proba_poisson_inferieur(i, k) for k in range(0, 30)]
    import matplotlib.pyplot as plt
    
    plt.bar(range(0, 30), tab2, color='r', alpha=0.8)
    plt.bar(range(0, 30), tab1, color='b')

    plt.title(f'Loi de Poisson avec λ = {i}')
    plt.xlabel('k')
    plt.ylabel('Probabilité')
    plt.legend(['P(X ≤ k)', 'P(X = k)'])
    
    plt.show()
```
